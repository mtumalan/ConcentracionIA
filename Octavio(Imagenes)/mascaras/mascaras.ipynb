```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import glob
```

```python
image_paths = glob.glob('/home/mtumalan/Desktop/Repos/Clases/ConcentracionIA/Octavio(Imagenes)/DL_Data_sets/images/SegmentationMasks/*_mask.png')

urban_land = np.array([0,255,255])
agriculture_land = np.array([255,255,0])
rangeland = np.array([255,0,255])
forest_land = np.array([0,255,0])
water = np.array([0,0,255])
barren_land = np.array([255,255,255])
unknown = np.array([0,0,0])

for path in image_paths:
    image = cv2.imread(path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (256, 256))

    image_copy = image.copy()

    mask = np.zeros(image.shape, dtype=np.uint8)

    f, ax = plt.subplots(1,8, figsize=(20, 10))
    ax[0].imshow(image)

    mask = cv2.inRange(image_copy, urban_land, urban_land)
    ax[1].imshow(mask)
    mask = cv2.inRange(image_copy, agriculture_land, agriculture_land)
    ax[2].imshow(mask)
    mask = cv2.inRange(image_copy, rangeland, rangeland)
    ax[3].imshow(mask)
    mask = cv2.inRange(image_copy, forest_land, forest_land)
    ax[4].imshow(mask)
    mask = cv2.inRange(image_copy, water, water)
    ax[5].imshow(mask)
    mask = cv2.inRange(image_copy, barren_land, barren_land)
    ax[6].imshow(mask)
    mask = cv2.inRange(image_copy, unknown, unknown)
    ax[7].imshow(mask)
```
